## Checking the October Data and Grouping based on Event Type

```python
 #Read October data (smaller file - ~4.2M rows)
oct_events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv",
                             header=True,
                             inferSchema=True)

oct_events.groupBy("event_type").count().orderBy("count", ascending=False).show()
#oct_events.display()
```

## Checking the Scheme of October Data

```python
# Display basic info
print(f"October 2019 - Total Events: {oct_events.count():,}")
print("\\n" + "="*60)
print("SCHEMA:")
print("="*60)
oct_events.printSchema()
```

## Checking the November Data

```python
nov_events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv",
                            header = True,
                            inferSchema = True)
            
nov_events.display()
```

# Filter the October events DataFrame to show only rows where event_type is "purchase"

```python
oct_events.filter("event_type = 'purchase'").display()
```

# Show the top 5 most purchased products from the October events data

```python
oct_Top5 = oct_events.groupBy("product_id").count().orderBy("count", ascending=False).limit(5).withColumnRenamed("count", "product_count")
oct_Top5.display()
oct_Top5.display()
```
